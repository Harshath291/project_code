```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('tips')
df
```

```python
df.info()
```

```python
df.isnull.sum()
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
df['smoker']=lb.fit_transform(df['smoker'])
df['sex']=lb.fit_transform(df['sex'])
df['day']=lb.fit_transform(df['day'])
df['time']=lb.fit_transform(df['time'])
```

```python
df.dtypes
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
sns.heatmap(np.abs(df.corr()),cmap='Purples',annot=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
sns.heatmap(np.abs(df.corr())>0.7,cmap='Purples',annot=True)
plt.show()
```

```python


import matplotlib.pyplot as plt
sns.heatmap(np.abs(df.corr())>0.5,cmap='Greens',annot=True)
plt.show()
```
